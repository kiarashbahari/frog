```r
data<-read.csv("data/hours1.csv", skip=1)
data<-data|>select(temp:expected_time)|>mutate(temp=as.factor(temp))
data<- data|>select(-embryos)
data
```

```r
ggplot(data, aes(x=time, y=expected_time, group=temp, color=temp))+
geom_point(alpha=0.5)
```

```r
abc23<-data|>filter(temp==23)
abc18<-data|>filter(temp==18)
lmm18<-lm(expected_time~  time+0,data=abc18)
lmm23<-lm(expected_time~  time+0,data=abc23)
summary(lmm18)
```

```r
summary(lmm23)
```

```r
y_pred <- predict(lmm18)

# Calculate the correlation between the predicted and actual values
cor.test(y_pred, abc18$expected_time)
```

```r
y_pred <- predict(lmm23)

# Calculate the correlation between the predicted and actual values
cor.test(y_pred, abc23$expected_time)
```

```r
ger<-data.frame(Temperature=c("18°C","23°C"),
              slope=c(0.6087622,1.0853567),
               SE=c(0.02359,0.01842))
ger
```

```r
Figure2<- ggplot(ger,aes(x=Temperature,y=slope,color=Temperature))+
geom_point(size=4)+
geom_errorbar(aes(x='18°C',ymin=0.6087622-SE18,ymax=0.6087622+SE18),width=0.2,color='red')+
geom_errorbar(aes(x='23°C',ymin=1.0853567-SE23,ymax=1.0853567+SE23),width=0.2)+
scale_y_continuous(limits=c(0.4,1.2),breaks=seq(0,1.3,0.05))+
labs(x='Incubation Temperature (Degree Celsius)',y='Development Rate',color='Incubation Temperature\n(Degree Celsius)')

Figure2
```

```r
slope1 <- 0.6087622
se1 <- 0.02359
n1 <- 46

slope2 <- 1.0853567
se2 <- 0.01842
n2 <- 40

t <- (slope1 - slope2) / sqrt(se1^2/n1 + se2^2/n2)
df <- n1 + n2 - 2
p_value <- 2 * pt(abs(t), df, lower.tail = FALSE)

ptable<-data.frame(t=t,
df=df,
p_value=p_value)
ptable
```
